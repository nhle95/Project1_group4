```python
#https://cfpub.epa.gov/ghgdata/inventoryexplorer/#transportation/entiresector/allgas/category/current

import pandas as pd 
import matplotlib.pyplot as plt
Texas = pd.read_csv("data/Texas_emissions.csv")
Texas
```

```python
texas_recent = Texas[['Texas Emissions, Transportation Sector, MMT CO2 eq.',"2015","2016","2017","2018","2019","2020"]]
texas_recent
total_texas_recent=texas_recent.loc[texas_recent['Texas Emissions, Transportation Sector, MMT CO2 eq.']=='Total',:]
total_texas_recent.columns =['Texas Emissions_Transportation Sector', '2015', '2016',
       '2017', '2018', '2019', '2020']
total_texas_recent
```

```python
total_texas_recent=total_texas_recent.set_index('Texas Emissions_Transportation Sector')
```

```python
new_df=total_texas_recent.T
```

```python
new_df.plot()
plt.xlabel("Year")
plt.ylabel('Metric Tons of Carbon Dioxide(CO2)')
plt.title('Texas Emissions from Transportation Sector')
plt.show()
```

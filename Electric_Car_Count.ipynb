```python
vehicleReg21=pd.read_excel(r'data/VehicleRegistrationByTypeCount.xlsx', sheet_name='2021')
vehicleReg20=pd.read_excel(r'data/VehicleRegistrationByTypeCount.xlsx', sheet_name='2020')#imports the excel file based on sheet named 2020
vehicleReg19=pd.read_excel(r'data/VehicleRegistrationByTypeCount.xlsx', sheet_name='2019')#imports the excel file based on sheet named 2019
vehicleReg18=pd.read_excel(r'data/VehicleRegistrationByTypeCount.xlsx', sheet_name='2018', header=1) #imports the excel file based on sheet named 2018
vehicleReg17=pd.read_excel(r'data/VehicleRegistrationByTypeCount.xlsx', sheet_name='2017', header=1) #header=1 skips first line and makes second line the headers
vehicleReg16=pd.read_excel(r'data/VehicleRegistrationByTypeCount.xlsx', sheet_name='2016', header=1)
```

```python
# reduced the columns to only EV, PHEV, and HEV registrations
vehicleReg21= vehicleReg21[['State', 'Electric (EV)', 'Plug-In Hybrid Electric (PHEV)','Hybrid Electric (HEV)']]
vehicleReg20= vehicleReg20[['State', 'Electric (EV)', 'Plug-In Hybrid Electric (PHEV)','Hybrid Electric (HEV)']]
vehicleReg19= vehicleReg19[['State', 'Electric (EV)', 'Plug-In Hybrid Electric (PHEV)','Hybrid Electric (HEV)']]
vehicleReg18= vehicleReg18[['State', 'Electric (EV)', 'Plug-In Hybrid Electric (PHEV)','Hybrid Electric (HEV)']]
vehicleReg17= vehicleReg17[['State', 'Electric (EV)', 'Plug-In Hybrid Electric (PHEV)','Hybrid Electric (HEV)']]
vehicleReg16= vehicleReg16[['State', 'Electric (EV)', 'Plug-In Hybrid Electric (PHEV)','Hybrid Electric (HEV)']]
#set index to states
vehicleReg21=vehicleReg21.set_index('State')
vehicleReg20=vehicleReg20.set_index('State')
vehicleReg19=vehicleReg19.set_index('State')
vehicleReg18=vehicleReg18.set_index('State')
vehicleReg17=vehicleReg17.set_index('State')
vehicleReg16=vehicleReg16.set_index('State') 
```

```python
vehicleReg21['Total']=vehicleReg21['Electric (EV)']+vehicleReg21['Plug-In Hybrid Electric (PHEV)']+vehicleReg21['Hybrid Electric (HEV)']
vehicleReg20['Total']=vehicleReg20['Electric (EV)']+vehicleReg20['Plug-In Hybrid Electric (PHEV)']+vehicleReg20['Hybrid Electric (HEV)']
vehicleReg19['Total']=vehicleReg19['Electric (EV)']+vehicleReg19['Plug-In Hybrid Electric (PHEV)']+vehicleReg19['Hybrid Electric (HEV)']
vehicleReg18['Total']=vehicleReg18['Electric (EV)']+vehicleReg18['Plug-In Hybrid Electric (PHEV)']+vehicleReg18['Hybrid Electric (HEV)']
vehicleReg17['Total']=vehicleReg17['Electric (EV)']+vehicleReg17['Plug-In Hybrid Electric (PHEV)']+vehicleReg17['Hybrid Electric (HEV)']
vehicleReg16['Total']=vehicleReg16['Electric (EV)']+vehicleReg16['Plug-In Hybrid Electric (PHEV)']+vehicleReg16['Hybrid Electric (HEV)']
vehicleReg16.tail(3) #adds the three columns together to make new totals column
```

```python
ranked_df=vehicleReg16.loc[:'Wyoming'].sort_values('Electric (EV)',ascending=False)
Ranked_States = ranked_df.index
type(Ranked_States)
```

```python
Index= vehicleReg16.index #stores index from df into variable 
```

```python
EV2016=vehicleReg16['Electric (EV)'].tolist() #converts to a list 
EV2017=vehicleReg17['Electric (EV)'].tolist()
EV2018=vehicleReg18['Electric (EV)'].tolist()
EV2019=vehicleReg19['Electric (EV)'].tolist()
EV2020=vehicleReg20['Electric (EV)'].tolist()
EV2021=vehicleReg21['Electric (EV)'].tolist()
```

```python
#plots the EV for multiple years
w=.1
X_axis=np.arange(len(Index))
X2_axis=[i+w for i in X_axis]
X3_axis = [i+w for i in X2_axis]
X4_axis = [i+w for i in X3_axis]
X5_axis = [i+w for i in X4_axis]
X6_axis = [i+w for i in X5_axis]
plt.figure(figsize=(15,7))

plt.bar(X_axis ,EV2016, w, label='2016')
plt.bar(X2_axis,EV2017, w, label='2017')
plt.bar(X3_axis,EV2018, w, label='2018')
plt.bar(X4_axis,EV2019, w, label='2019')
plt.bar(X5_axis,EV2020, w, label='2020')
plt.bar(X6_axis,EV2021, w, label='2021')

plt.xticks(X_axis+(5*w)/2, Index, rotation=45) # sets the ticks in center for the 6 years reported
plt.xlabel("States",size=15)
plt.ylabel("# of Registrations", size=15)
plt.title("Registrations of Electric Cars 2016-21",size=15)
plt.legend()
plt.ylim(0,85000)
plt.xlim(41.8,50.85)
plt.grid()
plt.savefig('output/Registrations of Electric Cars 2016-21.jpeg')
plt.show()
```

```python
US16total=vehicleReg16.iloc[-1,-1]
US17total=vehicleReg17.iloc[-1,-1]
US18total=vehicleReg18.iloc[-1,-1]
US19total=vehicleReg19.iloc[-1,-1]
US20total=vehicleReg20.iloc[-1,-1]
US21total=vehicleReg21.iloc[-1,-1]
Totals=[US16total,US17total,US18total,US19total,US20total,US21total]
Totals
Years=[2016,2017,2018,2019,2020,2021]
```

```python
# plt.bar(Years,Totals, color=['blue','yellow','red','purple','pink','cyan'])
sns.set_theme(style='darkgrid')
plots=sns.barplot(x=Years,y=Totals, edgecolor='black')
for bar in plots.patches:
    plots.annotate(format(bar.get_height(), '.0f'),
                   (bar.get_x() + bar.get_width() / 2,
                    bar.get_height()), ha='center', va='center',
                   size=11, xytext=(0,5),
                   textcoords='offset points')
plt.title('Number of Electric Cars in U.S. from 2016 to 2021',loc='right')
plt.xlabel('Years')
plt.ylabel('# of Regisrations  of Electric Cars\n(in millions)')
plt.ylim(0,8000000)
plt.savefig('output/Number of Electric Cars in U.S. from 2016 to 2021.jpeg')
plt.show()
```

```python
plt.rcdefaults()
fig, ax = plt.subplots()
ax.set_facecolor('Lavender')

ax.barh(Years, Totals, color='maroon',)
# for index, value in enumerate(Totals):
#     plt.text(value, index,
#              str(value))


ax.set_yticks(Years, labels=Years)
ax.set_xlabel('Number of Registrations \n (millions)')
ax.set_ylabel('Years', rotation=0)
ax.set_title('Electric Based Vehicle Trends in the Past 5 years')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.savefig('output/Electric Based Vehicle Trends in the Past 5 years.jpeg')
plt.show()
```
